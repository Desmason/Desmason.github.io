```python
import osmnx as ox
import pandas as pd

start_pos = (25.02603,121.55202)

tags = {'shop': True,'amenity':True}

G = ox.geometries.geometries_from_point(start_pos, tags, dist=2000)
```

```python
def get_shop(x , y):
    df = pd.DataFrame()
    a=0

    if ("shop" or "amenity") in G:
        for idx in range(len(G)):
            if G['amenity'][idx] == "restaurant":
                pt = G['geometry'][idx].centroid
                if 10**5*(ox.distance.euclidean_dist_vec(x, y,pt.y,pt.x)) < 200:
                    B = ox.bearing.calculate_bearing( x, y, pt.y, pt.x)
                    L = ox.distance.euclidean_dist_vec(x, y, pt.y, pt.x)
                    df = pd.concat([df, pd.DataFrame.from_records([{ 'Num': a, 'loc': pt, "name": G['name'][idx] ,"bearing": B,"length": L*10**5, "type":"restaurant" }])])
                    a=a+1    
            if G['amenity'][idx] == "cafe":
                pt = G['geometry'][idx].centroid
                if 10**5*(ox.distance.euclidean_dist_vec(x, y,pt.y,pt.x)) < 200:
                    B = ox.bearing.calculate_bearing( x, y, pt.y, pt.x)
                    L = ox.distance.euclidean_dist_vec(x, y, pt.y, pt.x)
                    df = pd.concat([df, pd.DataFrame.from_records([{ 'Num': a, 'loc': pt, "name": G['name'][idx] ,"bearing": B,"length": L*10**5, "type":"cafe" }])])
                    a=a+1 
    
                    
    return(df)
    


    
    
    
    #df = df.sort_values(by=['length'],ascending=False)
    
```

```python
get_shop(start_pos[0],start_pos[1])
```
